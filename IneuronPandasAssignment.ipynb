```python
import pandas as pd
```

```python
#Q1. How do you load a CSV file into a Pandas DataFrame?
                '''
                    We can load the data using read_csv() its a function in pandas.
                '''
data=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
#converting the csv file into a dadtaframe.
df=pd.DataFrame(data)
print(df.head())
```

```python
#Q2. How do you check the data type of a column in a Pandas DataFrame?
                #It can be done using the function called dtypes.
print(df.dtypes)
```

```python
#Q3. How do you select rows from a Pandas DataFrame based on a condition?
c=df[df['Age']>19]
#print(c)
#for multiple conditiion
mc=df.loc[(df['Age']>18) & (df['Age']<40)]
print(mc)
```

```python
#Q4. How do you rename columns in a Pandas DataFrame?
            #Using the function called rename
df.rename(columns={'Ticket':'Ticket Number'},inplace=True)
print(df.head())
```

```python
#Q5. How do you drop columns in a Pandas DataFrame?
            #we can drop a specific column using drop function and need to mention the axis =0(horizontal) ,axis=1(vertical)
rem=df.drop(['SibSp'],axis=1)
print(rem)
```

```python
#Q6. How do you find the unique values in a column of a Pandas DataFrame?
            #we can find the unique columns using "unique" dunction
df['Name'].unique() #will list all the unique name
```

```python
#Q7. How do you find the number of missing values in each column of a Pandas DataFrame?
                #we can find the number of missing values using 'isnull()' and to count the number of null values use 'sum()'
df.isnull().sum()
```
