```python
#Q10. How do you merge two Pandas DataFrames on a specific column?

# importing modules
import pandas as pd
# creating a dataframe
df1 = pd.DataFrame({'Name':['Raju', 'Rani', 'Geeta', 'Sita', 'Sohit'],
                    'Marks':[80, 90, 75, 88, 59]})
# creating another dataframe with different data
df2 = pd.DataFrame({'Name':['Raju', 'Divya', 'Geeta', 'Sita'],
                    'Grade':['A', 'A', 'B', 'A'],
                    'Rank':[3, 1, 4, 2 ],
                    'Gender':['Male', 'Female', 'Female', 'Female']})
# display df1
display(df1)
# display df2
display(df2)
#merge
print('Merging')
print(df1.merge(df2[['Name','Grade','Rank']]))
print()
print("joins")
print(df1.merge(df2[['Name','Grade']],on='Name',how='left'))#left represent the left join
print()
print("Merging all thr parameters")
print(df2.merge(df1[['Marks', 'Name']]))
```

```python
#Q11. How do you group data in a Pandas DataFrame by a specific column and apply an aggregation function?

import pandas as pd
technologies = {
    'Courses':["Spark","PySpark","Hadoop","Python","PySpark","Spark"],
    'Fee' :[20000,25000,26000,22000,24000,35000],
    'Duration':['30day','40days','35days','40days','60days','60days'],
    'Discount':[1000,2300,1200,2500,2000,2000]
              }
df = pd.DataFrame(technologies)
#print(df)

#groupby
g=df.groupby('Courses')['Fee','Discount',].aggregate('sum')
#g=df.groupby('Courses')['Fee','Discount',].aggregate('max')
#g=df.groupby('Courses')['Fee','Discount',].aggregate('min')
#gg=df.groupby('Courses').aggregate('sum') by default all the nnumberic values will be calculated
print(g)
```

```python
#Q12. How do you pivot a Pandas DataFrame?
        #pivot is used to reshape the a given dataframe by the index/column values
import pandas as pd
data=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/10_pivot/weather.csv")
df=pd.DataFrame(data)
print(df)
#in here index maens rows
#reshaping the date as rows and city as columns
z=df.pivot(index='date',columns='city')
print(z)
print()
#to get only the temperature
df.pivot(index='date',columns='city',values='temperature')
```

```python
#Q25. What is the difference between .loc and .iloc in Pandas?

#loc()
'''   -we have to pass the name of the row or column which we want to select.
      -loc() can accept the boolean data unlike iloc()'''
print(df.loc[3,'city']) #in the second parameter we need to give the label
print(df[0:4])
#iloc
'''    -we have to pass an integer index in the method to select a specific row/column
    -iloc() does not accept the boolean data unlike loc()'''
print(df.iloc[0:3,1:3]) #in iloc we need to give the seconf parameter only the index
```
